- from selenium.webdriver.common.by import By

- find_element(s)(By.ID, " "): id 속성을 사용하여 접근
- find_element(s)(By.CLASS_NAME, " "): 클래스를 사용하여 접근
- find_element(s)(By.NAME, " "): name 속성을 사용하여 접근
- find_element(s)(By.XPATH, " "): xpath 속성을 사용하여 접근
- find_element(s)(By.LINK_TEXT, " "): 앵커태그(a 태그)에 사용되는 텍스트로 접근 
- find_element(s)(By.PARTIAL_LINK_TEXT, " "): 앵커태그(a 태그)에 사용되는 일부 텍스트로 접근
- find_element(s)(By.TAG_NAME, " "): 태그를 사용하여 접근
- find_element(s)(By.CSS_SELECTOR, " "): CSS 선택자를 사용하여 접근<br><br>

```python
import time
import selenium
from selenium import webdriver

path = 'C:/cakd7/Tool/chromedriver.exe'
# driver.execute_script("alert('자바스크립트 코드 적용!!');") #자바 코드 실행
driver = webdriver.Chrome(path)
driver.get(url ='https://www.naver.com/')
element = driver.find_element(By.CLASS_NAME, 'link_login')
element.click()

idBox = driver.find_element(By.ID, 'id');
pwBox = driver.find_element(By.ID, 'pw');

n_id  = ''
n_pw = ''

driver.execute_script("arguments[0].value=arguments[1]", idBox,n_id); #빠르게 붙여넣어라
time.sleep(1)
driver.execute_script("arguments[0].value=arguments[1]", pwBox,n_pw);
# element.send_keys(n_pw) #입력해주기
time.sleep(2)
element = driver.find_element(By.CLASS_NAME, 'btn_login')
element.click()

time.sleep(2)
# 로그인 등록안함
element = driver.find_element(By.ID, 'new.dontsave')
element.click()

time.sleep(2)
#메일들어가기
element = driver.find_element(By.CLASS_NAME, 'ico_mail')
element.click()

#메일 목록
elements_title = driver.find_elements(By.CLASS_NAME,'mail_title')
elements_from = driver.find_elements(By.CSS_SELECTOR,'#list_for_view div.name._ccr\(lst\.from\) > a')
# for element in elements_from:
#     elements_from2=element.find_elements(By.TAG_NAME, 'a')

#예쁘게 출력
import pandas as pd
mail=pd.DataFrame(columns=['from','mail_title'])

i=0
for element in elements_title:
    mail.loc[i,'mail_title'] = element.text 
    i+=1

i=0
for element in elements_from:
    mail.loc[i,'from'] = element.text
    i+=1 
    
# # 종료
sleep(3)
driver.close()

```

```python
path = 'C:/cakd7/tool/chromedriver.exe'
url = 'http://naver.com'

# 1. 네이버 열기
driver = webdriver.Chrome(path)
driver.get(url)

# 2. 메일 창 열기
mail = driver.find_element(By.CSS_SELECTOR,'div.group_nav li.nav_item a.nav')
mail.click()
time.sleep(3)

# 3. 로그인하기
m_id = ''
m_pw = ''

### id
naver_id = driver.find_element(By.ID, 'id')
naver_id.click()
pyperclip.copy(m_id)
naver_id.send_keys(Keys.CONTROL, 'v')  # Keys.CONTROL:컨트롤 키 누르기 'v': v누르기 => ctrl + v: copy한 아이디 붙여넣기
time.sleep(3)

### password
naver_pw = driver.find_element(By.ID, 'pw')
naver_pw.click()
pyperclip.copy(m_pw)
naver_pw.send_keys(Keys.CONTROL, 'v') # Keys.CONTROL:컨트롤 키 누르기 'v': v누르기 => ctrl + v: copy한 아이디 붙여넣기
time.sleep(3)

###로그인 클릭
login_click = driver.find_element(By.ID, 'log.login')
login_click.click()
time.sleep(3)


# 4. 1페이지 메일 리스트 크롤링
who = driver.find_elements(By.CSS_SELECTOR,'#list_for_view div.name._ccr\(lst\.from\) > a')
header = driver.find_elements(By.CSS_SELECTOR, '#list_for_view div.subject > a._d2\(mcDragndrop\|html5DragStart\) > span > strong')

# 5.데이터프레임화
mail_df = pd.DataFrame(columns = ['who','contents'])

mail_who = []
mail_contents = []

for i in range(len(who)):
    mail_who.append(who[i].text)
    mail_contents.append(header[i].text)
    
mail_df['who'] = mail_who
mail_df['contents'] = mail_contents


driver.close()
mail_df
```
