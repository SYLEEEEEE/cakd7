```python
import numpy as np
import pandas as pd
import random
import time
import datetime
```

```python

np.random.seed(1)
df_c = pd.DataFrame(columns=['id','cusno','gender','age','region'])
df_c['id'] = random.sample(range(1,1001), 1000)
df_c['cusno'] = np.random.randint(1001, 1101, 1000)
df_c['gender'] = np.random.randint(0,2, 1000)
df_c['age'] = np.random.randint(0,81, 1000)
df_c['region'] = np.random.randint(1,11, 1000)
df_c.head()
```

```python
np.random.seed(1)
df_d = pd.DataFrame(columns=['product_20','product_21','price_20','price_21','quantity_20','quantity_21','amount_20','amount_21','sales'])
df_d['product_20'] = np.random.choice((10, 20, 30), 1000)
df_d['product_21'] = np.random.choice((10, 20, 30), 1000)
df_d['price_20'] = np.random.uniform(1000,50001, 1000)
df_d['price_21'] = np.random.uniform(1000,50001, 1000)
df_d['quantity_20'] = np.random.randint(1,101, 1000)
df_d['quantity_21'] = np.random.randint(1,101, 1000)
df_d['amount_20'] = df_d.price_20 * df_d.quantity_20
df_d['amount_21'] = df_d.price_21 * df_d.quantity_21
df_d['sales'] = np.where(df_d.amount_20<df_d.amount_21,1,0)
```

```python
df = df_c.join(df_d, how='inner')
df
```

```python
# 10세 미만은 구매를 안할 것을 ㅗ보고 10~80으로 데이터 변환
df['age'] = np.random.randint(10,81, 1000)
df
```

```python
1.지역에 따른 구매금액비교 >>>> 부유한 지역 확인1,2,3,4
2.나이(20대 30대, 40대, 50대)   (10살 미만은 잘라냄)와 성별로 구매금액 비교 >>> 가장 구매력 좋은 연령대와 성별 30대 여성 40대 여성, 가장 구매력이 낮은 60대 이상 남자
3.21년도와  매출하락과, 고가제품 구매율상승  >>경제상황 분석 빈익빈 부익부
4.제품 가격애 따른 
5.고가제품 구매 연령 2,30대 젊은층 큼, 저가제품(식료품) 주부 구매
```

```python
# 지역에 따른 구매금액 비교
#20년 21년 구매금액의 합계를 정렬해서 금액이 적은구간 부터 지역값을 100개씩 차례대로 대입함
df['amount_tt'] = df.amount_20 + df.amount_21 # 20년 21년 구매금액의 합계
df.amount_tt = np.sort(df.amount_tt)
for i in range(10):
    df.region[i*100:(i+1)*100] = 10-i
df
```

```python
# 나이대별 구매고객

def agec(x):
    if x < 30:
        return 20
    elif x < 40:
        return 30
    elif x < 50:
        return 40
    elif x < 60:
        return 50
    else:
        return 60
        
df['agec'] = df.age.apply(agec)
```
