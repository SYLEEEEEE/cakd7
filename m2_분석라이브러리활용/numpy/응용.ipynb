[numpy random 함수]

- np.random.seed : seed를 통한 난수 생성
- np.random.randint : 정수 난수 1개 생성
- np.random.rand : 0부터 1사이의 균일분포에서 난수 매트릭스 배열 생성
- np.random.randn : 가우시안 표준 정규분포에서 난수 매트릭스 배열 생성
- np.random.shuffle : 기존의 데이터의 순서 바꾸기
- np.random.choice : 기존 데이터에서 sampling


```python
# 파이썬 랜덤함수 Python random
import random
print(random.random(), '\n')  # [0,1) 사이의 난수
print(random.randint(1,10), '\n') # [1,10] 사이의 정수
print(random.uniform(10,20), '\n') # min, max 사이의 flost 
print(random.randrange(10), '\n') # 지정 범위 내 int
print(random.choice([1,2,3,4,5]), '\n') # 리스트 내부의 요소를 랜덤 선택
li = [1,2,3,4,5]
print(random.sample(li,3), '\n') # 리스트 요소를 중복이 안되게 출력
random.shuffle(li) # 리스트 요소를 다시 섞어서 리턴
print(li)
```
